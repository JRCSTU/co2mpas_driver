# Compute and plot Acceleration versus Velocity graph for each vehicle

```python
veh_ids = [35135, 27748, 15109, 8183, 26629, 8358, 17145, 17146, 35361, 5768, 3408, 15552, 8620, 8592, 5779, 8267, 
           4396, 4416, 34474, 9885, 7976, 34196, 34024, 8996]
v_des = [np.round(random.randint(45, 150) / 3.6, 2) for i in
             range(len(veh_ids))]
ds = [np.round(random.random(), 1) for i in range(len(veh_ids))] # random driver style values in the range (0, 1)
gs_style = [np.round(random.random(), 1) for i in range(len(veh_ids))] # random gear shifting values in the range (0, 1)
v_start = [random.randint(0, 10) for i in range(len(veh_ids))] # random starting values in the range (0, 10)
dt = 0.1
duration = 100
times = np.arange(0, duration, dt)
# Remember to Insert vehicle data base path (db_path)
vehicles = [driver(dict(vehicle_id=i, db_path='../co2mpas_driver/db/EuroSegmentCar.csv',
                            inputs=dict(inputs={'gear_shifting_style': gs_style[idx],
                                                'starting_velocity': v_start[idx],
                                                'driver_style': ds[idx],
                                                'desired_velocity': v_des[idx],
                                                'sim_start': 0,
                                                'sim_step': dt,
                                                'duration': duration}))) for idx, i in enumerate(veh_ids)]
```

```python
for idx, my_veh in enumerate(vehicles):
    outputs = my_veh['outputs'] # outputs of the dispatcher
    layout = go.Layout(title=f"Acceleration_Velocity_{veh_ids[idx]}_desired_velocity_{np.round(v_des[idx], 2)}_start_velocity_{np.round(v_start[idx], 2)}_driver_style_{ds[idx]}_gear_shifting_style_{gs_style[idx]}",
                xaxis=dict(title='velocity (in m/s)'),
                yaxis=dict(title='Acceleration (in m/s2)'))
    data = [dict(x=outputs['velocities'], y=outputs['accelerations'])] # select the desired outputs
    fig = go.Figure(data=data, layout=layout)
    iplot(fig)
```
